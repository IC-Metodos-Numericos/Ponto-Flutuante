```python
import numpy as np
import plotly.graph_objects as go
```

```python
# Valor inicial grande onde a adição de 1 começa a ser ignorada
z = np.float64(40**10)
z2 = z

step = 8

# Gerar 1000 floats consecutivos em float64 (1 ulp entre eles)
x = np.linspace(z , z+step , 100, dtype=np.float64)

# Avaliar a expressão (x + 1) - x
y = x + 1 - x


# Valor inicial grande onde a adição de 1 começa a ser ignorada


x2 = np.linspace(z2 , z2+step ,1000, dtype=np.float64)

# Avaliar a expressão (x + 1) - x
y2 = x2 + 1 - x2

# Plot
fig = go.Figure()
fig.add_trace(go.Scatter(x=x, y=y,
                         mode="lines+markers",
                         line=dict(color="red"),
                         marker=dict(size=3, color="blue"),
                         name="(x + 1) - x"))

fig.add_trace(go.Scatter(x=x2, y=y2,
                         mode="lines+markers",
                         line=dict(color="green"),
                         marker=dict(size=3, color="black"),
                         name="(x2 + 1) - x2",
                         line_shape="hv")
                         )

fig.update_layout(
    title="Salto de precisão: (x + 1) - x ao longo de floats consecutivos (float64)",
    xaxis_title="índice (1 ulp por passo)",
    yaxis_title="(x + 1) - x",
    xaxis=dict(tickformat=".16e") 
)

fig.show()
```

```python
import numpy as np
import plotly.graph_objects as go

# Valor grande onde a adição de 1 é ignorada
z = np.float64(40**10)
z2 = z
step = 8

# Gerar valores
x = np.linspace(z, z + step, 10, dtype=np.float64)
y = x + 1 - x

x2 = np.linspace(z2, z2 + step, 100, dtype=np.float64)
y2 = x2 + 1 - x2

# Índices como eixo x
indices1 = np.arange(len(x))
indices2 = np.arange(len(x2))

# Plot
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=indices1,
    y=y,
    mode="lines+markers",
    line=dict(color="red"),
    marker=dict(size=3, color="blue"),
    name="(x + 1) - x"
))

fig.add_trace(go.Scatter(
    x=indices2,
    y=y2,
    mode="lines+markers",
    line=dict(color="green"),
    marker=dict(size=3, color="black"),
    name="(x2 + 1) - x2",
    line_shape="hv"
))

fig.update_layout(
    title="Salto de precisão: (x + 1) - x ao longo dos índices (float64)",
    xaxis_title="índice dos pontos",
    yaxis_title="(x + 1) - x",
    xaxis=dict(tickmode="linear"),
    hovermode="x unified"
)

fig.show()
```

```python
x = 40**10
l = np.nextafter(2, 0)
print(f'{"l":<27} = {l}')
print(f'{"x == x + l":<27} = {x == x + l}')  # True, because x + l is indistinguishable from x
print(f'{"np.nextafter(x, np.inf) - x":<25} = {np.nextafter(x, np.inf) - x}')
```

```python
import numpy as np
import plotly.graph_objects as go

# Valor inicial grande onde a adio de 1 comea a ser ignorada
z = np.float64(40**10)

# Gerar 1000 floats consecutivos em float64 (1 ulp entre eles)
x = np.empty(1000, dtype=np.float64)
x[0] = z
for i in range(1, 1000):
    x[i] = np.nextafter(x[i-1], np.inf)

# Avaliar a expresso (x + 1) - x
y = x + 1 - x


# Valor inicial grande onde a adio de 1 comea a ser ignorada
z2 = np.float64(40**10)

x2 = np.empty(10000, dtype=np.float64)
x2[0] = z2
for i in range(1, 10000):
    x2[i] = np.nextafter(x2[i-1], np.inf)

# Avaliar a expresso (x + 1) - x
y2 = x2 + 1 - x2

# Plot
fig = go.Figure()
fig.add_trace(go.Scatter(x=x, y=y,
                         mode="lines+markers",
                         line=dict(color="red"),
                         marker=dict(size=3, color="blue"),
                         name="(x + 1) - x"))

fig.add_trace(go.Scatter(x=x2, y=y2,
                         mode="lines+markers",
                         line=dict(color="green"),
                         marker=dict(size=3, color="black"),
                         name="(x2 + 1) - x2",
                         line_shape="hv")
                         )

fig.update_layout(
    title="Salto de preciso: (x + 1) - x ao longo de floats consecutivos (float64)",
    xaxis_title="ndice (1 ulp por passo)",
    yaxis_title="(x + 1) - x",
    hovermode="x unified",
    xaxis=dict(tickformat=".e") 
)

fig.show()
```

```python
for xi, yi in zip(x, y):
    print(f"x = {xi:.17e}, --------- y = {yi}")
```

```python
z = np.float64(40**10)
ulp = np.nextafter(z, np.inf, dtype=np.float64) - z

x = np.linspace(z, z + 100, 1000, dtype=np.float64)
y = x + 1 - x

x2 = np.linspace(z, z + 100, 10000, dtype=np.float64)
y2 = x2 + 1 - x2

fig = go.Figure()

# Trace principal
fig.add_trace(go.Scatter(x=x, y=y,
                         mode="lines+markers",
                         line=dict(color="green"),
                         marker=dict(color="red", size=3),
                         name='(x + 1) - x'))

fig.add_trace(go.Scatter(x=x, y=y,
                         mode="lines+markers",
                         line=dict(color="red"),
                         marker=dict(color="blue", size=3),
                         name='(x + 1) - x'))

# Linha horizontal da ULP
fig.add_trace(go.Scatter(x=[x[0], x[-1]], y=[ulp, ulp],
                         mode="lines",
                         line=dict(dash='dash', color='green'),
                         name=f"ULP(z)  {ulp:.1e}"))

fig.update_layout(
    title="Anlise de (x + 1) - x e ULPS em ponto flutuante (float64)",
    xaxis_title="x",
    yaxis_title="(x + 1) - x",
    hovermode="x unified"
)

fig.show()
```
